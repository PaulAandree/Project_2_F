```python
import pandas as pd
```

Leemos los archivos

```python
Int_movil = pd.read_csv(r'data\Ingresos_Internet_Movil.csv')
Int_fijo = pd.read_csv(r'data\Ingresos_Internet_Fijo.csv')
```

Aplicamos Data Cleansing

```python
#Int_movil['Ingresos (miles de $)'] = Int_movil['Ingresos (miles de $)'].str.replace('[.$ ]','', regex=True)
Int_movil['Periodo'] = Int_movil['Periodo'].str.replace('\s+|\d{4}', '', regex=True)
Int_movil['Ingresos (miles de $)'] = Int_movil['Ingresos (miles de $)'].astype(int)

Int_fijo['Ingresos (miles de pesos)'] = Int_fijo['Ingresos (miles de pesos)'].str.replace('.','')
Int_fijo['Periodo'] = Int_fijo['Periodo'].str.replace('\s+|\d{4}', '', regex=True)
```

Guardamos los nuevos csv

```python
Int_movil.to_csv('data\Ingresos_Internet_Movil_CD.csv', index=False)
Int_fijo.to_csv('data\Ingresos_Internet_Fijo_CD.csv', index=False)
```

```python
# Leer los archivos CSV en DataFrames
df_ingresos = Int_movil
df_inflacion = pd.read_csv(r'data\df_inflacion.csv')

# Extraer el valor de la inflaci칩n trimestral
inflacion_trimestral = df_inflacion['Inflacion trimestral'].dropna().iloc[-1]

# Multiplicar la columna "Ingresos (miles de $)" por la inflaci칩n trimestral
df_ingresos['Ingresos (miles de $)'] *= inflacion_trimestral

# Guardar el resultado en un nuevo archivo CSV
df_ingresos.to_csv('archivo_ingresos_actualizado.csv', index=False)
```
